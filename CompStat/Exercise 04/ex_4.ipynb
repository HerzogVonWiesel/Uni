```python
import numpy as np
```

```python
# data
y = [50, 100]
x1 = [150, 200]
x2 = [1.5, 2]

# ordinary least squares solution
X = np.array([x1, x2]).T
# add column of ones for intercept
X = np.c_[np.ones(X.shape[0]), X]
beta = np.linalg.inv(X.T @ X) @ X.T @ y
print(beta)
```

```python
_lambda = 1
beta = np.linalg.inv(X.T @ X + _lambda*np.identity(3)) @ X.T @ y
print(beta)
```

```python
# Run ridge regression
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=1.0)
ridge.fit(X, y)
print(ridge.coef_)
print(ridge.intercept_)
```

```python
import pyreadr

result = pyreadr.read_r('model.rda')

# done! let's see what we got
model = result["model"] # extract the pandas data frame for object df1
print(model.head())

result = pyreadr.read_r('spec.rda')

# done! let's see what we got
spec = result["spec"] # extract the pandas data frame for object df1
print(spec.head())
```

```python
!pip install plotly
```

```python
# plot spec
import plotly.express as px
fig = px.line(spec, x="V1", y="V2", title='Spectrum')
fig.show()
```

```python
model_210 = model.iloc[:, 210]
model_215 = model.iloc[:, 215]
model_225 = model.iloc[:, 225]

# plot models
import plotly.graph_objects as go
fig = go.Figure()
fig.add_trace(go.Scatter(x=spec['V1'], y=spec['V2'], name='Spectrum'))
fig.add_trace(go.Scatter(x=model_210.index, y=model_210, name='210'))
fig.add_trace(go.Scatter(x=model_215.index, y=model_215, name='215'))
fig.add_trace(go.Scatter(x=model_225.index, y=model_225, name='225'))
fig.show()
```

```python
# run OLS to see each models contribution to the spectrum
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(model, spec['V2'])
print(reg.coef_)
print(reg.intercept_)
print(reg.score(model, spec['V2']))
```

```python
# rename model columns to their index
model.columns = range(model.shape[1])
print(model.head())
```

```python
# visualize coefficients
import plotly.graph_objects as go
fig = go.Figure()
fig.add_trace(go.Scatter(x=model.columns, y=reg.coef_, name='Coefficients'))
fig.show()
```

```python
# visualize regression and spectrum
import plotly.graph_objects as go
fig = go.Figure()
fig.add_trace(go.Scatter(x=spec['V1'], y=spec['V2'], name='Spectrum'))
fig.add_trace(go.Scatter(x=spec['V1'], y=reg.predict(model), name='Regression'))
```

```python
# Compute lasso regression
from sklearn.linear_model import LassoCV
lasso = LassoCV(cv=5, n_alphas=100)
lasso.fit(model, spec['V2'])
print(lasso.coef_)
print(lasso.intercept_)
print(lasso.score(model, spec['V2']))
```

```python
# Get the selected variables (models)
selected_variables = model.columns[lasso.coef_ != 0]

# Compare the results to the OLS solution
print("Selected variables (models) using Lasso:")
print(selected_variables)
```

```python
#visualize coefficients
import plotly.graph_objects as go
fig = go.Figure()
fig.add_trace(go.Scatter(x=model.columns, y=lasso.coef_, name='Coefficients'))
fig.show()
```

```python
# visualize regression and spectrum
import plotly.graph_objects as go
fig = go.Figure()
fig.add_trace(go.Scatter(x=spec['V1'], y=spec['V2'], name='Spectrum'))
fig.add_trace(go.Scatter(x=spec['V1'], y=lasso.predict(model), name='Regression'))
fig.show()
```
